The prediction is based mainly on the average score of best result in final in the two qualification events in Shanghai and Budapest. the weight that is given to every column during calculation of final prediction score is based on personal estimation. 


```python
# penalty - number of scores bellow fifty
scores_columns = ['shanghai-qual-run1','shanghai-qual-run2','shanghai-final-run1','shanghai-final-run2',

                  'budapest-qual-run1','budapest-qual-run2','budapest-final-run1','budapest-final-run2'
                  ]
df_paris['penalty'] = df_paris[scores_columns].apply(lambda x: (x<50).sum(), axis=1)
```

```python
# Average of best scores in final in both qualification events 
df_paris['avg'] = df_paris[['shanghai-final-best', 'budapest-final-best']].mean(axis=1)
```

```python
df_paris
```

```python
df_paris['avg'] = df_paris[['shanghai-final-best', 'budapest-final-best']].mean(axis=1)
```

```python
# Bonus for participating at Tokyo Olympics
df_paris['bonus'] = (df_paris['tokyo-participation'] == 'yes').astype(int)
```

```python
df_paris['prediction_score'] = df_paris['avg'] + df_paris['bonus'] - df_paris['penalty']
```
